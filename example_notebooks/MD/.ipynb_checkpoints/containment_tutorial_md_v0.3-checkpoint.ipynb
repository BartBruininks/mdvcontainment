## Visualization Wrappers

```python
def visualize_3D(name, command_string='-e visualization.vmd'):
    """
    Visualizes the GRO file using the included render settings.

    Feel free to change this function to whatever visualizer you prefer.
    """
    # Plot using VMD
    !vmd $name $command_string
    return
```

```python
def draw_containment_nx(G, images, write=True, fname='containment.png', imsize=0.2, lwidth=1, nodesize=0, dpi=300, prog='neato'):
    """
    Draws the containment graph using networkx and the provided images.

    """
    node_size = np.max(mpimg.imread(images[list(images.keys())[0]]).shape)*nodesize

    # The amount of images
    N = len(images)

    # Setting up the layout of the graph coordinates
    #pos=nx.spring_layout(G,k=k)
    pos = nx.nx_agraph.graphviz_layout(G, prog=prog)

    # draw with images on nodes
    nx.draw_networkx(G,pos,width=lwidth,edge_color="grey",alpha=1, node_size=node_size, node_color='white', arrows=False)
    ax=plt.gca()
    ax.axis('off')
    fig=plt.gcf()
    trans = ax.transData.transform
    trans2 = fig.transFigure.inverted().transform
    for n in G.nodes():
        (x,y) = pos[n]
        xx,yy = trans((x,y)) # figure coordinates
        xa,ya = trans2((xx,yy)) # axes coordinates
        a = plt.axes([xa-imsize/2.0,ya-imsize/2.0, imsize, imsize ])
        a.imshow(mpimg.imread(images[n]))
        a.set_aspect('equal')
        a.axis('off')
    if write:
        plt.savefig(fname, dpi=dpi, bbox_inches="tight")

    return
```

## 1 - Bicelle Containment

```python
# Load the GRO
base_name = 'bicelle'
u = mda.Universe('gros/bicelle.gro')
selection_string = 'name C2A C2B D2A D2B C3A C3B D3A D3B C4A C4B D4A D4B C5A C5B D5A D5B'
selection = u.select_atoms(selection_string)
```

```python
def get_component_composition(atomgroup, molar=False):
    """
    Returns the composition of an atomgroup as a dict {resname : count}.
    """
    if molar:
        composition = dict(zip(
            *np.unique(atomgroup.residues.resnames, return_counts=True)))
    else:
        composition = dict(zip(
            *np.unique(atomgroup.resnames, return_counts=True)))
    return composition
```

```python
def plot_composition(composition, labels=None, fontsize=20, radius=1, label_distance=1.3, write=False, fname='composition.png', dpi=600):
    """
    Plots a piechart of the composition dictionary {resname : count}.
    """
    # I dont know how to do this without this stupid function...
    def autopct_more_than_100(pct):
        return ('%1.f%%' % pct) if pct > 100 else ''
    
    # Creates the labels and sizes, allowing for more labels than sizes
    #  this can be used so the coloring in all the graphs is the same.
    if labels is None:
        labels = list(composition.keys())
    sizes = []
    for label in labels:
        try:
            sizes.append(composition[label])
        except KeyError:
            sizes.append(0)
    sizes = np.array(sizes, dtype=float)
    
    fig1, ax = plt.subplots()
    p, t, a = ax.pie(sizes, labels=None, textprops={'fontsize': fontsize}, radius = radius, labeldistance=label_distance, autopct=autopct_more_than_100)

    # normalize dataframe (not actually needed here, but for general case)
    normsizes = sizes/sizes.sum()*100
    # create handles and labels for legend, take only those where value is > 1
    h,l = zip(*[(h,lab) for h,lab, size in zip(p, labels, sizes) if size > 0])
    
    ax.legend(h, l,loc="best", bbox_to_anchor=(1,1), fontsize=fontsize)
    
    if write:
        plt.savefig(fname, dpi=dpi, bbox_inches="tight")
    return plt
```

```python
def plot_component_composition(component, render_img, composition_img, outsides=None, write=False, fname='component_composition.png', dpi=600):
    """
    Plots the component composition dual plot.
    """
    plt.subplot(1, 2, 1) # row 1, col 2 index 1
    plt.imshow(mpimg.imread(render_img))
    plt.axis('off')

    plt.subplot(1, 2, 2) # index 2
    plt.imshow(mpimg.imread(composition_img))
    plt.axis('off')
    plt.tight_layout()
    plt.title(f'{component}', fontsize=40)

    if outsides is not None:
        if component in outsides:
            plt.title(f'root {component}', fontsize=40)

    if write:
        plt.savefig(fname, dpi=dpi, bbox_inches="tight")

    return plt
```

```python
def write_all_component_composition(containers, selection, base_name='test', residue=False, dpi=600, residue_mask=True):
    """
    Plots the component composition figure for every component.

    Residue can be set to True to take to most common component ID per residue and project it on all atoms
    of the residue.

    Residue mask can be used to state that in the inverted atomgroup, no atoms are allowed which are part of
    a residue in the selection atomgroup (this is useful for when you make subselection of residues in the
    selection). For example only selecting the tail beads in a lipid, results in adding all lipid atoms to 
    the selection upon making the atomgroup. These atoms are added to the selection atomgroup as well. Meaning
    all atoms are counted once.

    This can be used instead of bare renders.
    """
    for component in containers.get_components():
        atomgroup = containers.get_atomgroup_from_components([component], residue=residue)
        # This completes the selection at the residue level. In principle 
        #  this causes the results to be much more intuitive when one uses
        #  subselections of residues.
        if residue_mask:
            if component < 0:
                atomgroup = atomgroup - selection.residues.atoms
            if component > 0:
                atomgroup = atomgroup | (atomgroup.residues.atoms & selection.residues.atoms)
        composition = get_component_composition(atomgroup)
        plot_composition(
            composition, 
            labels = np.unique(u.residues.resnames), 
            write = True, 
            fname = f'{base_name}_{component}_composition.png', 
            fontsize = 35, 
            dpi = dpi,
        )
        plot_component_composition(
            component, 
            containers._images[component], 
            f'{base_name}_{component}_composition.png', 
            outsides = containers.get_root_components(),
            write = True, 
            fname = f'{base_name}_component_composition_{component}.png', 
            dpi = dpi,
        )
```

```python
containers.load_renders(f'{base_name}_', ftype='png')
write_all_component_composition(containers, selection, base_name=base_name)
```

## Obtaining the content of a random voxel in a root (outside) component
This works for both the selection and the inveted selection.

1) Obtain all the additional data from the containers from the .data dict
2) Use the relabeled_combined_label_array for the component voxels ids.
3) Map the voxel id(s) back to the atom ids using the Voxels.voxel2atom dict.
4) Index the universe.atoms with the obtained atom ids.

```python
# Get all the voxel ids from the component labeled voxel grid.
target_component = containers.get_root_components()[0]
print(f'Node {target_component} has been selected as the outside.')
voxel_ids = np.array(np.where(containers.data['relabeled_combined_label_array'] == target_component))
```

```python
# Get the voxel object for all voxels.
voxels = containers.data['all_voxels']
```

```python
# Obtain the random voxel atom ids
selected_id = voxels.voxel2atom[tuple(voxel_ids.T[np.random.choice(voxel_ids.shape[1])])]
```

```python
# Show the atoms
u.atoms[selected_id].names
```

## Examples of upstream, downstream and children components

```python
target_components = [-1]
if target_components not in containers.get_root_components(): print(f'The parent of component {target_components} is {containers.get_upstream_components(target_components)[1]}.')
```

```python
# Get the composition of all downstream components of the target component(s).
target_components = [1]
downstream_components = containers.get_downstream_components(target_components)
downstream_atomgroup = containers.get_atomgroup_from_components(downstream_components)
downstream_composition = mdvc.containment.get_atomgroup_composition(downstream_atomgroup)
print(f'The downstream components of {target_components} are:\n{downstream_components}\n\nThe downstream composition of {target_components} is:\n{downstream_composition}')
```

```python
# Get the children of a component
target_components = [1, -1]
# Here we use the max depth to restrict the downstream search to one iteration (i.e. the children).
#  we also remove all the input target components using slicing for they are not a child.
target_components_children = containers.get_children_components(target_components)
print(f'The children of the target components {target_components} are:\n{target_components_children}')
```
