```python
import containment as mdvc
import MDAnalysis as mda
import networkx as nx
import numpy as np
```

```python
path = '../../example_notebooks/2D/nested_island_2D_rolled.gro'
u = mda.Universe(path)
selection = u.select_atoms('name A')
```

```python
containers = mdvc.Containers(selection, resolution=1, blur_amount=0)
```

```python
containers.plot()
```

```python
nx.draw_networkx(containers.data['bridges'].graph)
```

```python
nx.draw_networkx(containers.data['inv_bridges'].graph)
```

```python
all_bridges = containers.data['all_bridges'].graph
```

```python
nx.draw_networkx(all_bridges)
```

```python
all_nodes = set(all_bridges.nodes())
nx.draw_networkx(all_bridges.subgraph(all_nodes - set([3,6,2,5])))
```
